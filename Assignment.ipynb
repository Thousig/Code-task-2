```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from statsmodels.stats.contingency_tables import cochrans_q, mcnemar
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming X_scaled and Y are your feature matrix and labels
# Load data
filepath = "HR_data.csv"
data = pd.read_csv(filepath, sep=',')

# Encode categorical variables if necessary (assuming they are already numeric in your case)
label = LabelEncoder()
data['Cohort'] = label.fit_transform(data['Cohort'])
data['Phase'] = label.fit_transform(data['Phase'])
data['Round'] = label.fit_transform(data['Round'])
data['Frustrated'] = data['Frustrated'].apply(lambda x: 1 if x >= 3 else 0)

#data['HR_AUC'] = np.log(data['HR_AUC'])
#data['HR_Max'] = np.log(data['HR_Max'])
#data['HR_std'] = np.log(data['HR_std'])

# Split data into X (features) and Y (target)
Y = data['Frustrated']
X = data.drop('Frustrated', axis=1)

# Initialize StandardScaler
scaler = StandardScaler()

# Standardize the features
X_scaled = scaler.fit_transform(X)

# Initialize models
log_model = LogisticRegression(max_iter=10000, random_state=1)
rf_model = RandomForestClassifier(random_state=1)
neural_network = MLPClassifier(hidden_layer_sizes=(4, 4,4), max_iter=10000)

# Perform 5-fold stratified cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)

# Storage for predictions
preds_log = np.zeros_like(Y)
preds_rf = np.zeros_like(Y)
preds_nn = np.zeros_like(Y)

for train_index, test_index in cv.split(X_scaled, Y):
    # Train and predict with logistic regression
    log_model.fit(X_scaled[train_index], Y[train_index])
    preds_log[test_index] = log_model.predict(X_scaled[test_index])
    
    # Train and predict with random forest
    rf_model.fit(X_scaled[train_index], Y[train_index])
    preds_rf[test_index] = rf_model.predict(X_scaled[test_index])
    
    # Train and predict with neural network
    neural_network.fit(X_scaled[train_index], Y[train_index])
    preds_nn[test_index] = neural_network.predict(X_scaled[test_index])

# Calculate accuracy for each model
accuracy_log = accuracy_score(Y, preds_log)
accuracy_rf = accuracy_score(Y, preds_rf)
accuracy_nn = accuracy_score(Y, preds_nn)

print(f'Accuracy of Logistic Regression: {accuracy_log}')
print(f'Accuracy of Random Forest: {accuracy_rf}')
print(f'Accuracy of Neural Network: {accuracy_nn}')

# Generate confusion matrices for each model
cm_log = confusion_matrix(Y, preds_log)
cm_rf = confusion_matrix(Y, preds_rf)
cm_nn = confusion_matrix(Y, preds_nn)

# Plot confusion matrices
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.heatmap(cm_log, annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title('Confusion Matrix for Logistic Regression')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens', ax=axes[1])
axes[1].set_title('Confusion Matrix for Random Forest')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')

sns.heatmap(cm_nn, annot=True, fmt='d', cmap='Reds', ax=axes[2])
axes[2].set_title('Confusion Matrix for Neural Network')
axes[2].set_xlabel('Predicted')
axes[2].set_ylabel('Actual')

plt.tight_layout()
plt.show()

# Prepare data for Cochran's Q test
# Each row corresponds to a sample, and each column corresponds to a model
data = np.column_stack((Y, preds_log, preds_rf, preds_nn))

# Convert data to binary matrix (1 if prediction is correct, 0 otherwise)
correct_log = (data[:, 1] == data[:, 0]).astype(int)
correct_rf = (data[:, 2] == data[:, 0]).astype(int)
correct_nn = (data[:, 3] == data[:, 0]).astype(int)

# Combine the correct predictions
binary_matrix = np.column_stack((correct_log, correct_rf, correct_nn))

# Perform Cochran's Q test
q_result = cochrans_q(binary_matrix)
print(f'Cochran\'s Q statistic: {q_result.statistic}, p-value: {q_result.pvalue}')

# Interpret Cochran's Q test result
alpha = 0.05
if q_result.pvalue < alpha:
    print("Significant difference between the models (reject H0)")
else:
    print("No significant difference between the models (fail to reject H0)")

# Perform pairwise McNemar tests
def perform_mcnemar_test(preds_1, preds_2, labels):
    correct_1 = (preds_1 == labels).astype(int)
    correct_2 = (preds_2 == labels).astype(int)
    
    # Create contingency table
    table = np.zeros((2, 2))
    table[0, 0] = np.sum((correct_1 == 1) & (correct_2 == 1))
    table[0, 1] = np.sum((correct_1 == 1) & (correct_2 == 0))
    table[1, 0] = np.sum((correct_1 == 0) & (correct_2 == 1))
    table[1, 1] = np.sum((correct_1 == 0) & (correct_2 == 0))
    
    result = mcnemar(table, exact=True)
    return result.statistic, result.pvalue

# Logistic Regression vs Random Forest
statistic, pvalue = perform_mcnemar_test(preds_log, preds_rf, Y)
print(f'McNemar test between Logistic Regression and Random Forest: statistic={statistic}, p-value={pvalue}')

# Logistic Regression vs Neural Network
statistic, pvalue = perform_mcnemar_test(preds_log, preds_nn, Y)
print(f'McNemar test between Logistic Regression and Neural Network: statistic={statistic}, p-value={pvalue}')

# Random Forest vs Neural Network
statistic, pvalue = perform_mcnemar_test(preds_rf, preds_nn, Y)
print(f'McNemar test between Random Forest and Neural Network: statistic={statistic}, p-value={pvalue}')

```
